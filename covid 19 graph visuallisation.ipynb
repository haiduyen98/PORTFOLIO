```python
import numpy as np

df=pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv')
df.info()

```

```python
df['Confirmed']=df['Confirmed'].astype(float)
df['Deaths']=df['Deaths'].astype(float)
df['Recovered']=df['Recovered'].astype(float)
```

```python
df['Total_cases']=df[['Confirmed','Deaths','Recovered']].sum(axis=1)

df['Date'] = pd.to_datetime(df['Date'])
df['Year']=df['Date'].dt.year
df.info()
```

```python
df_us=df[df['Country']=='US'].groupby(by='Date').sum()
df_us
```

```python
df_worldwide=df.groupby(by=['Date']).sum()
df_worldwide
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Plotting worldwide cases
colors = ['blue', 'green', 'red','yellow']
w=df_worldwide.plot(figsize=(12,5), color=colors)
w.set_xlabel('Date')
w.set_ylabel('world wide cases')
w.title.set_text('Worldwide Covid Insights')
w.grid(True)
w.legend(loc='upper left')

```

```python
fig=plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
ax.plot(df_us['Total_cases'], label='US')
ax.plot(df_worldwide['Total_cases'], label='worldwide')
ax.set_xlabel('Date')
ax.set_ylabel('Number of total cases (confirmed+recover+deaths)')
ax.title.set_text('Worldwide covid insight')
ax.legend(loc='upper left')
ax.grid(True)


```

```python
us_df=df[df['Country']=='US'].groupby(by=['Date']).sum()

#us daily confirmed cases and daily deaths
us_df['Dailyconfirmed']= us_df['Confirmed'].sub(us_df['Confirmed'].shift())
us_df['Dailydeaths']= us_df['Deaths'].sub(us_df['Deaths'].shift())

fig=plt.figure(figsize=(20,8))
ax=fig.add_subplot(111)

us_df.reset_index(inplace=True)

ax.bar(us_df['Date'], us_df['Dailyconfirmed'], color='b',label='daily confirmed')
ax.bar(us_df['Date'], us_df['Dailydeaths'],color='r', label='deaths')
ax.set_xlabel('Date')
ax.set_ylabel('number')
ax.title.set_text('US daily confimred vs deaths')
plt.legend()
ax.grid(True)
```
