```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
color = ["green", "black", "Red"] 
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette=color)
```

```python
sns.histplot(tips['tip'],kde=True,bins=15)
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette=color)
sns.histplot(tips['tip'],kde=True,bins=15)
```

```python
sns.barplot(x="sex", y="tip", data=tips, palette=['red', 'yellow'])
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex", palette=color)
```

```python
sns.stripplot(x="day",y="tip",data=tips, hue="sex",palette=color,dodge=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True,cmap="YlGnBu")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",cmap="YlGnBu")
```

```python
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

df
```

```python
# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```
